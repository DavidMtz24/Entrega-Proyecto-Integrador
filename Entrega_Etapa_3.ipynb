### Situación Problema

El Programa para la Evaluación Internacional de Alumnos de la OCDE (Organización para la Cooperación y el Desarrollo Económico) es una encuesta que se aplica cada 3 años a alumnos de todo el mundo de 15 años. El propósito de la evaluación es calificar que tan preparados están los alumnos para poder convivir y participar plenamente en su sociedad, enfocándose en las áreas de lectura, matemáticas y ciencias. 

Gracias a los resultados obtenidos durante la prueba de OCDE en el año 2018, se pudo observar que los estudiantes mexicanos obtuvieron, en promedio, un puntaje menor que el promedio global de todos los países participantes de la OCDE. Además, en cuestión de porcentajes que describen el nivel de competencia en las áreas como lectura y matemáticas, los estudiantes mexicanos tienen un porcentaje alto de competencias bajas en las áreas de matemáticas y lectura. También cabe mencionar las tendencias de los resultados de los estudiantes mexicanos en comparación con las tendencias de los alumnos en promedio a nivel mundial. Y, al observar las tendencias, se observa que no hay un crecimiento o aumento significativo en los resultados de los estudiantes mexicanos para poder alcanzar a los del promedio de la OCDE.

Es por esto que una solución es diseñar un programa computacional que esté al servicio de los estudiantes y que les ayude a poder resolver sus dudas y obtener más competencias en las áreas en las que se les pone a prueba. Este programa debería de incluir varios elementos. El primero sería una sección dedicada a estudios de cada área, lectura, matemáticas y ciencias, en donde los alumnos puedan ingresar a leer e investigar sobre estas áreas para aprender. El siguiente elemento debería ser un grupo de preguntas, relacionadas a las áreas a evaluar, con las que el alumno pueda practicar y mejorar su desempeño en la prueba. También, un elemento que sería útil sería una forma en la que, alumnos que ya hayan presentado la prueba puedan ingresar preguntas del tipo que se ven en la prueba, o también profesores que conozcan qué tipo de preguntas se hacen. De esta forma, los alumnos que van a llevar a cabo la prueba se pueden familiarizar con las preguntas, y su desempeño mejorará gracias al programa.



```python
preguntas = []
import numpy as np
contadorp = 0
puntajes = []
cont_quiz = 0
cont_corr = 0
cont_incorr = 0 
import random
nombre = "preguntas"

def registrar_pregunta():
  print(f"Bienvenido a la sección para introducir preguntas.")
  leer_archivo(preguntas)
  id_pregunta = str(input("Ingrese el número de identificación de la pregunta: "))
  enfoque = str(input("Ingrese el enfoque de la pregunta: "))
  pregunta = str(input("Ingrese la pregunta: "))

  respuestas = []
  respuesta = ""

  for num in range(4):
    resp = str(input("Ingresa la opción de respuesta número " + str(num + 1) + ": "))
    respuesta = respuesta + "Respuesta " + str(num + 1) + ": " + resp + ", "
  
  respuestas.append(respuesta)
  
  respuesta_correcta = str(input("Ingrese la respuesta correcta: "))

  pregunta = [id_pregunta, enfoque, pregunta, respuestas, respuesta_correcta]
  
  print(f"Pregunta ingresada exitosamente\nID de la pregunta: {id_pregunta}\nEnfoque de la pregunta: {enfoque}\nPregunta: {pregunta}\nRespuestas: {respuestas}\nRespuesta correcta: {respuesta_correcta}")
  
  preguntas.append(pregunta)

  global contadorp

  contadorp = contadorp + 1

  print("La cantidad de preguntas son:", contadorp)

  global nombre

  grabar_archivo(nombre, preguntas)

  

def actualizar_pregunta():
  
  global preguntas

  preguntas = leer_archivo(preguntas)


  while True: 
    editar = int(input("Ingresa el ID de la pregunta a editar: "))
    
    if editar > contadorp:
      print("No existe pregunta con esa identificación")
      break

    elif editar <= contadorp:
      pae = preguntas[editar - 1]
      print(f"La pregunta que seleccionaste para editar es:\n{pae}")
      sera_editada = str(input("¿Deseas editar esta pregunta? "))

      if sera_editada != "":
        print("Ingrese la pregunta de reemplazo")
        id_pregunta_r = str(input("Ingrese el número de identificación de la nueva pregunta: "))
        enfoque_r = str(input("Ingrese el enfoque de la nueva pregunta: "))
        pregunta_r = str(input("Ingrese la pregunta: "))

        respuestas_r = []
        respuesta_r = ""

        for num in range(4):
          resp_r = str(input("Ingresa la opción de respuesta número " + str(num + 1) + ": "))
          respuesta_r = respuesta_r + "Respuesta " + str(num + 1) + ": " + resp_r + ", "
  
        respuestas_r.append(respuesta_r)
  
        respuesta_correcta_r = str(input("Ingrese la respuesta correcta: "))

        pregunta_reemplazo = [id_pregunta_r, enfoque_r, pregunta_r, respuestas_r, respuesta_correcta_r]

        preguntas[editar - 1] = pregunta_reemplazo

        print("Pregunta reemplazada exitosamente")
        print(f"Las preguntas son:\n{np.array(preguntas)}")
        grabar_archivo(nombre, preguntas)
        break

      else:
        print("La pregunta no se editará")
        break
    


def estudiar_area():
  leer_archivo(preguntas)
  area = input("Ingresa el área que quieres estudiar: ")
  preguntas_area = []
  for i in range(len(preguntas)):
    if area == preguntas[i][1]:
      preguntas_area.append(preguntas[i])

  print(f"Las preguntas a estudiar son:\n{np.array(preguntas_area)}")



def presentar_quiz():
  import time

  start = time.time()
  calificacion = 0

  for i in range(10):
    n = random.randint(0, contadorp)
    for j in range(4):
      preg = preguntas[n][j]
      print(preg)
    res = input("Ingresa la respuesta que crees correcta: ")
  
    if res == preguntas[i][4]:
      print("Respuesta correcta")
      calificacion = calificacion + 10
      cont_corr = cont_corr + 1
      
    elif res != preguntas[i][4]:
      print("Respuesta incorrecta")
      cont_incorr = cont_incorr + 1
  
  end = time.time()

  tiempo_total = (end - start)

  if tiempo_total > 600:
    print("Te pasaste del tiempo límite")
  
  if tiempo_total <= 600:
    print(f"Tu calificación es {calificacion}")
    puntajes.append(calificacion)

    global cont_quiz

    cont_quiz = cont_quiz + 1



def reportar_calificaciones():
  suma_punt = 0
  for i in range(len(puntajes)):
    suma_punt = suma_punt + puntajes[i]
  
  promedio = suma_punt / cont_quiz

  global cont_corr
  global cont_incorr

  print(f"Total de usuarios: {cont_quiz}\nPromedio de calificaciones: {promedio}\nPreguntas correctas: {cont_corr}\nPreguntas incorrectas: {cont_incorr}")

  grabar_archivo(nombre, promedio)


def salir():
  print("Gracias por utilizar este programa")



def menu():

  print("1. Registrar pregunta nueva")
  print("2. Actualizar preguntas de la prueba")
  print("3. Estudiar preguntas de cierta área")
  print("4. Presentar un quiz")
  print("5. Reporte de calificaciones")
  print("6. Salir")

  seleccion = int(input("Seleccione una opción introduciendo el número de la opción: "))

  return seleccion



def grabar_archivo(nombre, lista):
  with open("preguntas.txt","w") as nombre:
    for elemento in lista:
      nombre.writelines("%s\n" % elemento)



def leer_archivo(nombre):
  with open("preguntas.txt","r") as archivo:
    nombre = archivo.readlines()

    print(f"{nombre}")
    return nombre



def main():
  
  with open("preguntas.txt","w") as archivo:
    archivo.write("Lista de preguntas: ")

  while True:

    seleccion = menu()

    if seleccion != 1 and seleccion != 2 and seleccion != 3 and seleccion != 4 and seleccion != 5 and seleccion != 6:
      print("Opción incorrecta")

    elif seleccion == 1:
      registrar_pregunta()

    elif seleccion == 2:
      actualizar_pregunta()

    elif seleccion == 3:
      estudiar_area()

    elif seleccion == 4:
      presentar_quiz()

    elif seleccion == 5:
      reportar_calificaciones()
    
    elif seleccion == 6:
      salir()
      break
    


main()
```
